```python
print("Prime Number Checker")

choice = "y"

while choice.lower() == "y":
    num = int(input("\nPlease enter an integer between 1 and 5000: "))

    factors = []

    for i in range(1, num + 1):
        if num % i == 0:
            factors.append(i)

    if len(factors) == 2:
        print(f"{num} is a prime number.")
    else:
        print(f"{num} is NOT a prime number.")
        print(f"It has {len(factors)} factors:", end=" ")
        for f in factors:
            print(f, end=" ")
        print()

    choice = input("\nTry again? (y/n): ")
```

```python
print("Contact Manager")

contacts = [
    ["Guido van Rossum", "guido@python.org", "+1 555 123 4567"],
    ["Eric Idle", "eric@ericidle.com", "+44 20 7946 0958"]
]

def display_menu():
    print("\nCOMMAND MENU")
    print("list - Display all contacts")
    print("view - View a contact")
    print("add  - Add a contact")
    print("del  - Delete a contact")
    print("exit - Exit program")

display_menu()

while True:
    command = input("\nCommand: ").lower()

    if command == "list":
        for i, contact in enumerate(contacts, start=1):
            print(f"{i}. {contact[0]}")

    elif command == "view":
        number = int(input("Number: "))
        contact = contacts[number - 1]
        print(f"Name: {contact[0]}")
        print(f"Email: {contact[1]}")
        print(f"Phone: {contact[2]}")

    elif command == "add":
        name = input("Name: ")
        email = input("Email: ")
        phone = input("Phone: ")
        contacts.append([name, email, phone])
        print("Contact added.")

    elif command == "del":
        number = int(input("Number: "))
        contacts.pop(number - 1)
        print("Contact deleted.")

    elif command == "exit":
        print("Bye!")
        break

    else:
        print("Invalid command.")
```
